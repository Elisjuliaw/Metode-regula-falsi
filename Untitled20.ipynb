```python
# Representasi Pohon Keluarga
family_tree = {
    "Hadi": {"spouse": "Julia", "children": ["Bayu", "Desi", "Elis", "Rina", "Ardi"]},
    "Julia": {"spouse": "Hadi", "children": ["Bayu", "Desi", "Elis", "Rina", "Ardi"]},
    "Bayu": {"parents": ["Hadi", "Julia"], "children": ["Fahrul"]},
    "Desi": {"parents": ["Hadi", "Julia"], "children": ["Tari", "Nurul", "Yanto"]},
    "Elis": {"parents": ["Hadi", "Julia"], "children": ["Gunawan"]},
    "Rina": {"parents": ["Hadi", "Julia"], "children": ["Hamzah"]},
    "Ardi": {"parents": ["Hadi", "Julia"], "spouse": "Ana", "children": ["Eka", "Mira", "Bastian"]},
    "Ana": {"spouse": "Ardi", "children": ["Eka", "Mira", "Bastian"]},
    "Mira": {"parents": ["Ardi", "Ana"], "children": ["Anggun", "Boy"]},
    # Anak-anak dan cucu
    "Fahrul": {"parents": ["Bayu"], "children": []},
    "Tari": {"parents": ["Desi"], "children": []},
    "Nurul": {"parents": ["Desi"], "children": []},
    "Yanto": {"parents": ["Desi"], "children": []},
    "Gunawan": {"parents": ["Elis"], "children": []},
    "Hamzah": {"parents": ["Rina"], "children": []},
    "Eka": {"parents": ["Ardi", "Ana"], "children": []},
    "Bastian": {"parents": ["Ardi", "Ana"], "children": []},
    "Anggun": {"parents": ["Mira"], "children": []},
    "Boy": {"parents": ["Mira"], "children": []},
}

# Fungsi relasi
def get_parents(person):
    return family_tree.get(person, {}).get("parents", [])

def get_children(person):
    return family_tree.get(person, {}).get("children", [])

def get_spouse(person):
    return family_tree.get(person, {}).get("spouse", None)

def get_siblings(person):
    parents = get_parents(person)
    siblings = []
    for parent in parents:
        siblings.extend([child for child in get_children(parent) if child != person])
    return siblings

def get_brothers(person):
    return [sibling for sibling in get_siblings(person) if sibling in family_tree]

def get_sisters(person):
    return [sibling for sibling in get_siblings(person) if sibling in family_tree]

def get_aunts_uncles(person):
    parents = get_parents(person)
    aunts_uncles = []
    for parent in parents:
        aunts_uncles.extend(get_siblings(parent))
    return aunts_uncles

def get_grandparents(person):
    parents = get_parents(person)
    grandparents = []
    for parent in parents:
        grandparents.extend(get_parents(parent))
    return grandparents

def get_cousins(person):
    aunts_uncles = get_aunts_uncles(person)
    cousins = []
    for aunt_uncle in aunts_uncles:
        cousins.extend(get_children(aunt_uncle))
    return cousins

# Fungsi utama untuk mendapatkan semua relasi
def get_all_relations(person):
    relations = {
        "Orang Tua": get_parents(person),
        "Saudara Laki-Laki": get_brothers(person),
        "Saudara Perempuan": get_sisters(person),
        "Paman/Bibi": get_aunts_uncles(person),
        "Kakek/Nenek": get_grandparents(person),
        "Sepupu": get_cousins(person),
    }
    return relations

# Contoh penggunaan
people = ["Desi", "Elis", "Rina", "Ardi", "Fahrul", "Tari", "Nurul", "Yanto", "Hamzah", "Eka", "Mira", "Bastian", "Wanda", "Aji", "Gunawan", "Anggun", "Boy", "Ana"]
for person in people:
    print(f"\nRelasi untuk {person}:")
    relations = get_all_relations(person)
    for relation, names in relations.items():
        print(f"{relation}: {', '.join(names) if names else 'Tidak ada'}")
```
